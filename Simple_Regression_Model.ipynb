```python
# simple linear regression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import dataset
dataset =  pd.read_csv('C://Users//hp//JupiterNobook_Files_For_ML//Datasets//Salary_data.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
# since we donot have any missng or nan values t so no need to do data cleaning
# moving on to create train n test data 
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X, y, test_size = 0.2, random_state = 1  )
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
# predicted salares and y_test contains actual salaries
y_pred = regressor.predict(X_test)
```

```python
# visualize the training result
plt.scatter(X_train,y_train, color= 'red')
plt.plot(X_train,regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience')
plt.xlabel('years of expereince')
plt.ylabel('Salary')
plt.show()
```

```python
# visualize the test result
plt.scatter(X_test,y_test, color= 'red')
plt.scatter(X_test,y_pred, color= 'yellow')
plt.plot(X_train,regressor.predict(X_train), color='blue') # we need same regression blue line so we need not to replace training set by tet set
plt.title('Salary vs Experience')
plt.xlabel('years of expereince')
plt.ylabel('Salary')
plt.show()
```
