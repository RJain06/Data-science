```python
# importing the libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the database
dataset = pd.read_csv('C://Users//hp//JupiterNobook_Files_For_ML//Datasets//Data.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(X)
print(y)
```

```python
# taking care of missing data usng SI class we can replace missing salaries with average 
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan,strategy = 'mean')
# fit function helps to compute  using imputer class
imputer.fit(X[:,1:3])
# now we ahve to transform our data into readable data
X[:,1:3] = imputer.transform(X[:,1:3])
```

```python
# encoding categorical data (Why - as if we have some data which are similar like names of countries then we need to change them in some rder machine can understand thats why we are dong categorical data change
```

```python
# encoding the independant variables
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder',OneHotEncoder(),[0])], remainder ='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

```python
# encoding the dependent varable
from sklearn.preprocessing  import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

```python
# Splitting the data into trainng and test set
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X, y, test_size = 0.2, random_state = 1  )
```

```python
# Feature Scaling -  when to apply before or after    
# we need to apply it after splitting   as our test set is something whch s raw and ths is a technique to apply n get Sd n mean for all values on test set.

#Main techniques Standardisation and normalisation
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,3:] = sc.fit_transform(X_train[:,3:])
X_test[:,3:] = sc.fit_transform(X_test[:,3:])
```
